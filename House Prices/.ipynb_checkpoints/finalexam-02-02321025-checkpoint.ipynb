## 교차검증

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV
import optuna
```

```python
kfold = KFold(n_splits=8, shuffle=True, random_state=134)
val_score = []
for i, (tr_idx, val_idx) in enumerate(kfold.split(X)):

    # 1. 폴드별 훈련용과 검증용 데이터로 나누기
    X_tr, X_val, y_tr, y_val = X.iloc[tr_idx], X.iloc[val_idx], y.iloc[tr_idx], y.iloc[val_idx]

    # 2. 훈련하기
    model = XGBRegressor()
    model.fit(X_tr, y_tr)

    # 3. 검증 및 채점하기
    y_pred = model.predict(X_val)
    score  = np.sqrt(mean_squared_error(y_val, y_pred))
    val_score.append(score)
    
    print(f"{i+1} fold : {score:10.5f}")
```

```python
np.mean(val_score)
```

```python
def objective(trial):
    
    hyper_params = {
        "n_estimators": trial.suggest_int("n_estimators", 200, 600),
        "max_depth": trial.suggest_int("max_depth", 1, 15),
        "learning_rate": trial.suggest_float("learning_rate", 1e-6, 1, log=True),  
        "subsample": trial.suggest_float("subsample", 0.1, 1.0, log=True),  
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.1, 1.0, log=True), 
        "gamma": trial.suggest_float("gamma", 1e-9, 1.0, log=True),
        "min_child_weight": trial.suggest_int("min_child_weight", 1, 10),
        "reg_lambda": trial.suggest_float("reg_lambda", 1e-8, 1.0, log=True),
        "reg_alpha": trial.suggest_float("reg_alpha", 1e-8, 1.0, log=True),
    }
    
    kfold = KFold(n_splits=8, shuffle=True, random_state=134)
    val_scores = []
    for tr_idx, val_idx in kfold.split(X):

        # 1. 폴드별 훈련용과 검증용 데이터로 나누기
        X_tr, X_val, y_tr, y_val = X.iloc[tr_idx], X.iloc[val_idx], y.iloc[tr_idx], y.iloc[val_idx]

        # 2. 훈련하기
        model = XGBRegressor(**hyper_params)
        model.fit(X_tr, y_tr)

        # 3. 검증 및 채점하기
        y_pred = model.predict(X_val)
        score  = np.sqrt(mean_squared_error(y_val, y_pred))
        val_scores.append(score)

    return np.mean(val_scores)
```

## Optimization

```python
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=100)
```

## History

```python
optuna.visualization.plot_optimization_history(study)
```

```python
best_params = study.best_params
best_params
```
